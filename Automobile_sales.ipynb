```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Auto Sales data.csv',encoding='unicode_escape')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from datetime import datetime

temp_date = datetime(2020, 6, 1)  # Define the specific date
df['DAYS_SINCE_LASTORDER'] = (temp_date- df['ORDERDATE']).dt.days
df
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

df['year'] = df['ORDERDATE'].dt.year
result = df.groupby('year')['SALES'].sum().reset_index()

print(result)
```

```python
df[['PRICEEACH','PRICEEACH','MSRP']].describe()
```

```python
ax=sns.countplot(x='STATUS',data=df,hue='STATUS',palette='dark')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
plt.title('Distribution of STATUS')
plt.xlabel('STATUS')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

af=sns.countplot(x='PRODUCTLINE',data=df,hue='PRODUCTLINE',palette="coolwarm")


for container in af.containers:
    af.bar_label(container, fmt='%.0f')

plt.title('Distribution of PRODUCTLINE')
plt.xlabel('PRODUCTLINE')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
grouped=df.groupby('COUNTRY')['SALES'].sum()

top15=grouped.sort_values(ascending=False).head(15)

sales_values = top15.values
country_labels = top15.index

plt.pie(sales_values,labels=country_labels, autopct='%1.1f%%')
plt.title('Top 15 countries on sales contribution')
plt.show()
```

```python
sns.histplot(df['SALES'])
plt.title('Distribution of SALES')
plt.xlabel('SALES')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
grouped=df.groupby('CITY')['SALES'].sum()

top10 = grouped.sort_values(ascending=False).head(10)

sales_values = top10.values
country_labels = top10.index

colors = sns.color_palette("Set2")

plt.pie(sales_values,labels=country_labels, autopct='%1.1f%%',colors=colors)
plt.title('Top10 contibuting cities to sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
ab=df.groupby('CUSTOMERNAME')['SALES'].sum()

data1=ab.sort_values(ascending=False).head(5).reset_index()


gp=sns.barplot(x='CUSTOMERNAME',data=data1,y='SALES',hue='CUSTOMERNAME',palette='Spectral')


for container in gp.containers:
    gp.bar_label(container, fmt='%.0f')

plt.title('Top 5 Customers by Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='DEALSIZE',y='SALES',data=df,hue='DEALSIZE')
plt.title('Distribution by dealsize')
plt.show()
```

```python

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['year'] = df['ORDERDATE'].dt.year
result = df.groupby('year')['SALES'].sum().reset_index()

sns.lineplot(x='year',y='SALES',data=result)
plt.show()
```

```python
df["SALES (K)"] = df["SALES"] / 1000
ax=df.groupby(['PRODUCTLINE','DEALSIZE'])['SALES'].sum().reset_index()

sns.barplot(x='PRODUCTLINE',y='SALES',data=ax,hue='DEALSIZE')
plt.title('Relation between productline and dealsize along sales')
plt.show()
```

**Key Insights from Analysis**


1.United States leads in sales, followed by Spain and France. Notable sales from Australia and Singapore in the Eastern Hemisphere.

2.Among cities, Madrid tops sales, with San Diego and New York City following.

3.Classic Car parts generate the maximum revenue through sales.

4.Higher sales and customer activity in the mid months of 2019 compared to previous years.

5.Most deals are of medium and small size, with transactions typically containing 20 to 50 items.

6.Euro Shopping Channel and Mini Gifts Distribution Ltd. are best customers (highest spending, most frequent, and recent).
